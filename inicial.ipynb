# Python Insights - Analisando Dados com Python

### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
# Passo a passo
# Passo 1: Importar a base de dados
import pandas as pd

tabela = pd.read_csv("cancelamentos_sample.csv")





```

```python
# Passo 2: Visualizar a base de dados(entender a base + indentificar problemas)

tabela = tabela.drop(columns="CustomerID")

display(tabela)
```

```python
# Passo 3: Corrigir os problemas da base de dados(tratamento de dados)
# Valores vazios
display(tabela.info())

tabela = tabela.dropna()

display(tabela.info())
```

```python
# Passo 4: Análise da causa de cancelamento dos clientes
# Contar os valores da coluna cancelou

display(tabela["cancelou"].value_counts())
display(tabela["tempo_como_cliente"].mean())
display(tabela["cancelou"].value_counts(normalize = True))
```

```python
# Passo 5: Analise a causa de cancelamento dos clientes
# Comparar outras colunas com a coluna de cancelamento

import plotly.express as px

# 1º Criar o grafico

grafico = px.histogram(tabela, x = "assinatura", color = "cancelou", text_auto = True)


# 2º exibir o grafico
grafico.show()

grafico = px.histogram(tabela, x = "duracao_contrato", color = "cancelou", text_auto = True)

grafico.show()

grafico = px.histogram(tabela, x = "tempo_como_cliente", color = "cancelou", text_auto = True)

grafico.show()
```
